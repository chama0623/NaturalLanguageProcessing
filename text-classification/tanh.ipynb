# tanh関数とその微分のプロット

tanh関数は次で定義される. tanh関数はNNにおける活性化関数として用いられる.
$$y = tanh(x)\\
= \frac{e^x-e^{-x}}{e^x+e^{-x}}
$$

tanh関数
$$
\frac{dy}{dx} = \frac{(e^x+e^{-x})^2 -(e^x-e^{-x})^2}{(e^x+e^{-x})^2} \\
= \frac{(e^x+e^{-x})^2}{(e^x+e^{-x})^2} - \frac{(e^x-e^{-x})^2}{(e^x+e^{-x})^2}\\
= 1- tanh^2(x)\\
= 1-y^2
$$

```python
x = np.linspace(-5,5)
tanh = np.tanh(x)
tanhd = 1-tanh**2

plt.figure(facecolor="white")
plt.plot(x,tanh,label="tanh")
plt.plot(x,tanhd,label="$dy/dx$")
plt.legend(fontsize=13)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
